```python
import matplotlib.pyplot as plt

# Definindo os pontos de um triângulo no espaço (Coordenadas X e Y)
# Imagine isso no seu mapa mental: cada ponto é uma conexão!
x = [3, 1, 1, 3, 3]
y = [3, 3, 1, 1, 3]

plt.figure(figsize=(4, 4))
plt.plot(x, y, marker='.', linestyle='-', color='#0000FF')

# Adicionando títulos para exercitar a percepção
plt.title("Visualizando a Inteligência Espacial com Python")
plt.xlabel("Eixo Horizontal (X)")
plt.ylabel("Eixo Vertical (Y)")
plt.grid(True)

plt.show()
```
