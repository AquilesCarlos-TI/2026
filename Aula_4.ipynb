Fazer um programa que recebe 4 notas do aluo e calcula a média aritimética e imprimi a média.


```python
matematica= float(input("Quanto tirou em matemática?"))#entrada
portugues= float(input("Quanto você tirou em português?"))#entrada
historia= float(input("Quanto você tirou em história?"))#entrada
fisica= float(input("Quanto você tirou em física?"))#entrada

media= ((matematica + portugues + historia + fisica) / 4)#processamento

if(media >= 6):
  print ("Parabéns, voce passou e tirou :", (media))
elif(media >=5):
    print("Passou por pouco e tirou :", media)
else:
   print("Você reprovou e tirou :", media)
```
